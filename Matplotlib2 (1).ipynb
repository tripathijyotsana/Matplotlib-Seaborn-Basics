```python
import matplotlib.pyplot as plt
```

```python
#Line plot
x = [1,2,3,4,5]
y = [10,20,25,30,40]
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.title('This is our first line plot')

plt.show()
```

```python
#Canvas
plt.figure(figsize = (6,4))

x  = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

#line 1 plot
plt.plot(x,y1,label = "Line1", color = "red")
plt.plot(x,y2,label = "line2", color = "pink")

plt.legend()
plt.grid()
```

```python
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4,5]
Y1 = [1,4,9,16,25]
Y2 = [1,2,3,4,5]

plt.subplot(2,2,1)
plt.plot(X,Y1)

plt.subplot(2,2,4)
plt.plot(X,Y2)
```

```python
#sample data
x = [1,2,3,4,5]
y = [2,4,6,8,10]
```

```python
fig = plt.figure(figsize = (6,4))
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(x,y)
plt.title("sample plot")


ax2 = fig.add_axes([0.2,0.5,0.3,0.3])
ax2.plot(y,x)
plt.title("plot")
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [1,3,5,7,9]

plt.plot(x,y1,label = "axis1", color = "pink", lw=2.5)
plt.plot(x,y2, label = 'axis2', color = "black" , ls = '-.', marker = '*')

plt.title('Graph')
plt.legend()
```

```python
x = [1,2,3,4,5]
y = [5,7,4,6,8]
```

```python
#Scatter plot -> 2 columns/variables/features
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#Histplot -> shows spread of only one column/variable
import numpy as np
nums = np.random.randn(100)
plt.hist(nums,bins=20)
```

```python
#Box Plot -> Shows outliers of a single column
plt.boxplot(nums)
#min
#25% -> lower quartile
#50% -> Median
#75% -> Upper quartile
#max
"""
values between lower and upper quartile are generic or actual values
values outside of lower and upper quartile are considered as outliers.
"""
```
