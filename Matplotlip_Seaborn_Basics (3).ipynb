```python
x = np.linspace(0,5,11)
x
y = x ** 2
y
```

```python
plt.plot(x,y)
```

Subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

Object-Oriented-Way

```python
#Matplotlib allows the aspect ratio, DPI and figure size to specified when the figure object is created. You can use the figsize and dpi keyword arguements.
#figsize is the tuple of a width and height of the figure in inches and DPI is the dots-per-inches (pixel per inches).
```

```python
fig = plt.figure(figsize = (12,8), dpi = 500)
axis1 = fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(y,x)
```

Saving plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = x**2


fig = plt.figure()

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x, y)


ax.set_title("Basic Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")


plt.savefig("basicplot.png")


plt.show()
```

Working with images

```python
import matplotlib.image as mpimg
img = mpimg.imread("basicplot.png")
plt.imshow(img)
```
